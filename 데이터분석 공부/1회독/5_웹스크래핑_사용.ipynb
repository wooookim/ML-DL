```python
'''
1. 온라인서점 검색 결과 페이지 url 생성
2. requests.get() 함수로 검색 결과 페이지 html 가져오기
3. 뷰티풀수프로 html 파싱
4. 뷰티풀수프 find() 매서드로 a태그를 찾아 상세 페이지 url 추출
5. requests.get() 함수로 다시 도서 상세 페이지 html 가져오기
6. 뷰티풀수프로 html 파싱
7. 뷰티풀수프 find() 매서드로 '품목정보' div 태그 찾기
8. 뷰티풀수프 find_all() 매서드로 '쪽수'가 들어있는 tr 태그 찾기
9. 앞에서 찾은 테이블 행에서 get_text() 매서드로 td태그에 들어잇는 쪽수 가져오기
'''
def get_page_cnt(isbn):
  # 도서검색 페이지 url
  url = 'https://www.yes24.com/Product/Search?domain=BOOK&query={}'

  # url에 isbn을 넣고 html을 가져오기
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')

  # 검색 결과에서 해당 도서를 선택
  prd_info = soup.find('a', attrs = {'class':'gd_name'})

  # 도서 상세 페이지 가져오기
  url = 'https://www.yes24.com' + prd_info['href']
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')

  # 품목 정보 div 선택
  prd_detail = soup.find('div', attrs = {'id':'infoset_specific'})

  # 테이블에 있는 tr 태그 가져오기
  prd_tr_list = prd_detail.find_all('tr')

  # 쪽수가 들어있는 th 를 찾아 td에 담긴 값을 반환
  for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
      return tr.find('td').get_text().split()[0]
  return ''
```

```python
get_page_cnt(9791190090018)
```

```python
top10_books = books.head(10)
```

```python
def get_page_cnt2(row):
  isbn = row['isbn13']
  return get_page_cnt(isbn)
```

```python
'''
apply()
- 첫 번째 매개변수 : 실행할 함수
- 데이터프레임행 또는 열에 함수를 일괄 적용
- axis = 0 -> 열
- axis = 1 -> 행
'''
page_count = top10_books.apply(lambda row: get_page_cnt(row['isbn13']), axis = 1)
print(page_count)
```
